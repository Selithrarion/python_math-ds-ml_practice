```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
measurments = np.random.normal(loc=20, scale=5, size=150)
stats.probplot(measurments, dist='norm', plot=plt)
plt.show()
```

```python
import seaborn as sns
```

```python
ax = sns.boxplot(x=measurments)
```

```python
ax = sns.displot(x=measurments)
```

shapiro-wilk test

```python
from scipy.stats import shapiro
stat, p = shapiro(measurments)
print('stat=%.3f, p=%.3f\n'% (stat,p))
```

d'agustino's k-squared test

```python
from scipy.stats import normaltest
stat, p = normaltest(measurments)
print('stat=%.3f, p=%.3f\n'% (stat, p))
```
