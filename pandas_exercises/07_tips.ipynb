```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

sns.set_style('white')
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
df = pd.read_csv(url)
df.head()
```

```python
del df['Unnamed: 0']
```

```python
sns.displot(df.total_bill)
```

```python
sns.jointplot(x='total_bill', y='tip', data=df)
```

```python
sns.pairplot(df)
```

```python
sns.scatterplot(x='total_bill', y='tip', data=df, size='size')
```

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

```python
sns.stripplot(x='day', y='total_bill', data=df)
```

```python
sns.scatterplot(x='tip', y='day', data=df, hue='sex')
```

```python
sns.swarmplot(x='tip', y='day', data=df, hue='sex', size=4)
```

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=df)
```

```python
g = sns.FacetGrid(df, col='time')
g.map(sns.histplot, 'tip')
```

```python
g = sns.FacetGrid(df, col='sex', hue='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip', alpha=0.7)
g.add_legend()
```
