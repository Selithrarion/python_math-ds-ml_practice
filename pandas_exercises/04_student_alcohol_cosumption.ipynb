```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(url)
df.head()
```

```python
dfs = df.loc[:, 'school':'guardian']
dfs.head()
```

```python
cap = lambda x: x.upper()
```

```python
feats = ['Mjob', 'Fjob']
for feat in feats:
    dfs[feat] = dfs[feat].apply(cap)
```

```python
dfs.tail()
```

```python
def majority(dfs):
    dfs['legal_drinker'] = dfs.age > 17
majority(dfs)
dfs.head()
```
