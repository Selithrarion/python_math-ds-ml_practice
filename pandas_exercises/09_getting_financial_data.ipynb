```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import datetime as dt
```

```python
start = dt.datetime(2015,1,1)
end = dt.datetime.today()
start, end
```

```python
df_apple = web.DataReader(
    name='AAPL',
    data_source='av-daily',
    start=start,
    end=end,
    api_key='QRY1E4P8N91DQ3HQ'
)
df_apple.tail()
```

```python
df_apple['stock'] = 'AAPL'
df_apple.head()
```

```python
df_tesla = web.DataReader(
    name='TSLA',
    data_source='av-daily',
    start=start,
    end=end,
    api_key='QRY1E4P8N91DQ3HQ'
)
df_tesla['stock'] = 'TSLA'
df_tesla.tail()

df_ibm = web.DataReader(
    name='IBM',
    data_source='av-daily',
    start=start,
    end=end,
    api_key='QRY1E4P8N91DQ3HQ'
)
df_ibm['stock'] = 'IBM'

df_microsoft = web.DataReader(
    name='MSFT',
    data_source='av-daily',
    start=start,
    end=end,
    api_key='QRY1E4P8N91DQ3HQ'
)
df_microsoft ['stock'] = 'MSFT'
```

```python
df = pd.concat([df_apple, df_tesla, df_ibm, df_microsoft])
df
```

```python
df.set_index('stock', append=True, inplace=True)
df.head()
```

```python
vol = pd.DataFrame(df.volume)
vol.head()
```

```python
date = vol.index.get_level_values(0)
date = pd.DatetimeIndex(date)

vol['week'] = date.isocalendar().week.values
vol['year'] = date.year

pd.pivot_table(
    vol, 
    values='volume',
    index=['year', 'week'], 
    columns=['stock'], 
    aggfunc=np.sum
)
```

```python
vol_2015 = vol[vol.year == 2015]
pd.pivot_table(vol_2015, values='volume', index=['year'], columns=['stock'], aggfunc=np.sum)
```
